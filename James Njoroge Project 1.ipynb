#### Calculating Percentage of missing values per Data Set

```python
#Calculating the percentage of missing values per column in the data sets
def missing_values(data):
    "Function to identify data with missing values"
    #Identify total missing values per column
    #Sort the total missing values in order
    missing = data.isnull().sum().sort_values(ascending = False)

    #Calculating percentage of missing values
    percentage_missing = (data.isnull().sum() / len(data)).sort_values(ascending = False)

    #Store in a data frame
    missing_data = pd.DataFrame({"Missing Values":missing, "Percentage": percentage_missing}).reset_index()

    return missing_data
#Applying functions to data sets
missing_data = missing_values(df_title_basics)
missing_data
```
